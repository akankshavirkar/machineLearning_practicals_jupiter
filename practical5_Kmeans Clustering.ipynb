```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data.Age,data['BloodPressure'])
plt.xlabel('Age')
plt.ylabel('BloodPressure')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(data[['Age','BloodPressure']])
y_predicted
```

```python
data['cluster']=y_predicted
data.head()
```

```python
km.cluster_centers_

data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
plt.scatter(data1.Age,data1['BloodPressure'],color='Green')
plt.scatter(data2.Age,data2['BloodPressure'],color='Blue')
plt.scatter(data3.Age,data3['BloodPressure'],color='Orange')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='red',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('BloodPressure')
plt.legend()
```

```python
#Elbow method
```

```python
x1 = data.iloc[:,[8,3]].values
```

```python
from sklearn.cluster import KMeans
wcss_list=[]
```

```python
for i in range(1,16):
    kmeans = KMeans(n_clusters = i ,init = 'k-means++', random_state = 42)
    kmeans.fit(x1)
    wcss_list.append(kmeans.inertia_)
    
plt.plot(range(1,16),wcss_list)
plt.title('Elbow Method Graph')
plt.xlabel('Numbers of clusters (k)')
plt.ylabel('wcss_list')
plt.show
```
