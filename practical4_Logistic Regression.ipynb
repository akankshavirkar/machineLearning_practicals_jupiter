```python
# split independent and dependent vairable
x=df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
```

```python
x
```

```python
y=df.Outcome
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameters={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regression=GridSearchCV(classifier,param_grid=parameters,scoring='accuracy',cv=5)
```

```python
classifier_regression.fit(x_train,y_train)
```

```python
print(classifier_regression.best_params_)
```

```python
print(classifier_regression.best_score_)
```

```python
#prediction
y_pred=classifier_regression.predict(x_test)
```

```python
y_pred
```

```python
#accuracy score
from sklearn.metrics import accuracy_score,classification_report
```

```python
score= accuracy_score(y_pred,y_test)
```

```python
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
